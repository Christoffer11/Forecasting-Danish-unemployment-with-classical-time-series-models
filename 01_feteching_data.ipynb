# Fetching data

The registered unemployment is fetched from statbank of the national statistics bureau Statistics Denmark. 

## Importing packages and libraries

```python
import pandas as pd
import numpy as np
import requests
```

## Connecting to DST

```python
url = "https://api.statbank.dk/v1/data/AUS09/JSONSTAT?lang=en"

params = {
    "YDELSESTYPE": "TOT",
    "SAESONFAK": "22",
    "ContentsCode": "AUS09",
    "Tid": "*",
    "timeOrder": "Descending"
}

response = requests.get(url, params=params, timeout=30)
#response.raise_for_status()

data = response.json()
```

## Saving data

```python
dataset = data["dataset"]

time_index = dataset["dimension"]["Tid"]["category"]["index"]

periods = [
    period
    for period, position in sorted(
        time_index.items(),
        key=lambda item: item[1]
    )
]

df = pd.DataFrame({
    "date": pd.to_datetime(periods, format="%YM%m"),
    "unemployment_rate": dataset["value"]
})

df.to_csv("./data/unemployment.csv", sep=",", index=False)
```
